# Project Charter




## Business background

* O produto se destina a ambientes que necessitam de temperatura agradavel para a produção intelectual (escritórios, salas de lazer, sala de reuniões e etc).
* O produto tem a ambição de tentar equilibrar a temperatuda do ambiente com a estação que o ano se encontra.



## Scope
* O escopo do projeto é um ecossistema IOT capaz de receber dados pelos sensores e processar os dados e através de atuadores em ar-condicionados tentar equilibrar a temperatura.
* O usuário não precisará inputar dados manuais apenas instalar sensores de temperaturas no espaço.
